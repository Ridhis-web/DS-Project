```python
import pandas as pd
import numpy as np
import seaborn as sns
from statistics import mean, median, mode, variance, stdev
data = {'Student': ['Amit', 'Riya', 'Sohan', 'Neha', 'Rahul', 'Anjali', 'Ravi', 'Priya', 'Rohan', 'Meena', 'Rahul'],
'Marks': [78, 56, 90, 45, 66, 56, 90, 66, 78, 45, 56]}
df = pd.DataFrame(data)
marks = df['Marks']
mean_val = mean(marks)
median_val = median(marks)
mode_val = mode(marks)
variance_val = variance(marks)
stdev_val = stdev(marks)
print("Statistical Summary:")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
print(f"Variance: {variance_val:.2f}")
print(f"Standard Deviation: {stdev_val:.2f}")
```

```python
#Visualizations
# Histogram
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.histplot(marks, bins=5, kde=True, color='skyblue')
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
